```python
import numpy as np 
from numpy import random as npr
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from datetime import datetime as dt
import os
from sklearn import *
```

```python
def sigmoid(z):
    return((1 + np.exp(-z))**(-1))

def regularize(m):
    return(m/np.sum(m))

def ReLU(Z):
    return np.maximum(Z, 0)

def weighted_cross_entropy(y, p, m):
    return -sum(m*(y*np.log(p) + (1-y)*np.log(1-p)))

def pFromX(X, seed = 123):
    npr.seed(seed)
    T = X @ npr.randn(D, 7)/1.5
    T += sigmoid(T)
    T -= np.exp(T/5)
    T = T @ npr.randn(T.shape[1], 4)/1.5
    T -= np.tile(T.mean(0), T.shape[0]).reshape(T.shape)
    T /= np.tile(T.std(0), T.shape[0]).reshape(T.shape)/(10)
    p = sigmoid(T.mean(1) - T.mean())
    return p

```

```python
class NN():
    
    def __init__(self, name = None):
        self.name = name
        
    def generate_data(self, N, D, pFromX, seed = 123):
        npr.seed(seed)
        self.N, self.D = N, D
        self.m = np.repeat(1/N, N)
        mu = npr.randn(D) # means
        Sigma = np.eye(D) # covariance
        self.X = npr.multivariate_normal(mu, Sigma, size = N) # get data (X)
        self.latent_p = pFromX(self.X) # get latent P
        self.y = npr.binomial(1, self.latent_p, N) # true outcome values
      
    def add_data(self, X, pFromX, seed = 123):
        npr.seed(seed)
        self.N, self.D = X.shape
        self.m = np.repeat(1/self.N, self.N)
        self.X = X
        self.latent_p = pFromX(X)
        self.y = npr.binomial(1, self.latent_p, N)
        
    def get_output(self):
        # X = self.X[batch], B1 = self.B1[batch], etc. 
        self.Z1 = (self.X @ self.W1) + self.B1
        self.A = ReLU(self.Z1)
        self.z2 = (self.A @ self.w2) + self.b2
        self.p = sigmoid(self.z2)
        self.L = weighted_cross_entropy(self.y, self.p, self.m)
        
    def get_derivatives(self, batch):
        # partials (let B be batch size)
        # self.get_output(batch)
        B = len(batch)
        dpdz2 = np.diag(sigmoid(self.z2[batch])*(1-sigmoid(self.z2[batch]))) # (B, B)
        dz2db2 = np.ones(B) # B
        dz2dw2 = self.A[batch] # (B, h)
        dz2dA = np.tensordot(np.eye(B), self.w2, 0) # (B, B, h)

        # full derivatives of loss function
        self.dLdp = self.m[batch]*(-self.y[batch]/self.p[batch] + (1-self.y[batch])/(1-self.p[batch])) # B
        self.dLdz2 = self.dLdp @ dpdz2 # B 
        self.dLdb2 = (self.dLdz2 @ dz2db2).sum() # 1 (equivalent to sum(dLdz2))
        self.dLdw2 = self.dLdz2 @ dz2dw2 # h
        self.dLdA = np.tensordot(self.dLdz2, dz2dA, 1) # (B, h)
        self.dLdZ1 = self.dLdA * (self.Z1[batch] > 0) # (B, h)
        self.dLdB1 = self.dLdZ1.sum(0) # h
        self.dLdW1 = self.X[batch].T @ self.dLdZ1 # (D, h)
        
    def update_parameters(self):
        self.W1 -= self.lr*self.dLdW1
        self.B1 -= self.lr*self.dLdB1
        self.w2 -= self.lr*self.dLdw2
        self.b2 -= self.lr*self.dLdb2
        
    def train(self, h, lr, epochs, batch_size):
        # add parameters
        self.h = h 
        self.lr = lr
        self.epochs = epochs 
        self.batch_size = batch_size
        
        # instantiate parameters randomly and get outputs
        self.W1 = npr.randn(self.D*self.h).reshape((self.D, self.h))
        self.B1 = np.tile(npr.randn(self.h), self.N).reshape((self.N, self.h))
        self.w2 = npr.randn(self.h)
        self.b2 = npr.randn(1)[0]
        self.get_output()
        
        # loop
        self.losses = []
        n_batches = self.N // self.batch_size
        for epoch in range(epochs):
            batch_indices = np.split(npr.permutation(np.arange(self.N)), n_batches)
            for i, batch in enumerate(batch_indices):
                self.get_derivatives(batch)
                self.update_parameters()
                self.get_output()
            self.losses.append(self.L)
            
    def plot_losses(self):
        fig, ax = plt.subplots(figsize = (5, 4))
        sns.lineplot(np.arange(self.epochs), self.losses)
        ax.set(xlabel = 'Iterations', ylabel = 'Loss');
        
    def plot_ps(self):
        fig, ax = plt.subplots(figsize = (5, 4))
        sns.scatterplot(self.latent_p, self.p)
        sns.lineplot(np.arange(2), np.arange(2))
        ax.set(xlabel = 'latent p', ylabel = 'p');        

N, D = 100, 4
h, lr, epochs, batch_size = 3, 0.01, 5000, 10
nn = NN()
nn.generate_data(N, D, pFromX)
nn.train(h, lr, epochs, batch_size)
nn.plot_losses()

```
